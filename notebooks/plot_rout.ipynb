```python
def create_vectors(yaws_degrees, start_point=False, negate=False):

    if start_point:
        yaws_degrees = yaws_degrees - yaws_degrees[start_point]

    if negate: 
        yaws_degrees = 360  - np.array(yaws_degrees) 
    radians = np.deg2rad(yaws_degrees)
    hos = np.cos(radians)
    mot = np.sin(radians)

    vectors = np.stack((hos,mot), axis=-1)
    return vectors, yaws_degrees
```

```python

def plot_route_with_yaws(vectors, refrence_frame=0): 
    plt.figure(figsize=(10,15))

    plt.scatter(x_offsets[::3],y_offsets[::3], 1)

    plt.scatter(x_offsets[-1300:-200:3],y_offsets[-1300:-200:3], 1, color="red")

    plt.scatter(x_offsets[refrence_frame], y_offsets[refrence_frame])

    plt.quiver(x_offsets[1::20],y_offsets[1::20], vectors[::20,1], vectors[::20,0])
```
